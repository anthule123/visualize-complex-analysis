```python
import matplotlib.pyplot as plt

import numpy as np

from matplotlib.collections import PolyCollection
import random
fig, ax = plt.subplots(figsize = (10,10))

poly_set = np.zeros((2,3,2))
for i in range(0,2):
    for j in range(0,3):
        for k in range(0,2):
            poly_set[i,j,k] = random.random()*2

ax.add_collection(PolyCollection(poly_set,facecolors = 'green', linewidths=1))

### limits, title, annotation ###
M = 5
ax.set_xlim(-M,M)
ax.set_ylim(-M,M)
xvec = np.linspace(-5.,5.,100)                               
x,y = np.meshgrid(xvec, xvec)

plt.axhline(0, color='white')
plt.axvline(0, color='white')

plt.show()

```

```python
def rotate(x,y, r, phi): 
    r_old = np.sqrt(x**2+y**2)
    r_old *=r
    phi_old = 0
    if(x == 0):
        if(np.sign(y)==1):
            phi_old = np.pi/2
        elif(np.sign(y)==-1):
            phi_old = -np.pi/2
    else:
        phi_old = np.arctan(y/x)
    phi_old += phi
    return np.array([r_old*np.cos(phi_old), r_old*np.sin(phi_old)])
```

```python
poly_new = poly_set.copy()
r = 3
phi = random.random()*np.pi
for l in range(0,2):
    for i in range(0,3):
        poly_new[l][i] = rotate(poly_new[l][i][0],poly_new[l][i][1],r,phi)
```

```python
fig, ax = plt.subplots(figsize = (10,10))
ax.add_collection(PolyCollection(poly_set,facecolors = 'green', linewidths=1))
ax.annotate(3, poly_set[0][2])
ax.add_collection(PolyCollection(poly_new,facecolors = 'red', linewidths=1))

### limits, title, annotation ###
M = 6
ax.set_xlim(-M,M)
ax.set_ylim(-M,M)
xvec = np.linspace(-5.,5.,100)                               
x,y = np.meshgrid(xvec, xvec)
plt.plot([0, r*np.cos(phi)], [0, r*np.sin(phi)], marker ='v')
plt.axhline(0, color='white')
plt.axvline(0, color='white')

plt.show()
```
