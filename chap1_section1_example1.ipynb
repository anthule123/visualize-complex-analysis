```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#(1+ i) ^13
r = [np.sqrt(2)]
phi = [np.pi/4]
x=[0]
y=[0]
x.append(r[0]*np.cos(phi[0]))
y.append(r[0]*np.sin(phi[0]))

N =12
fig,ax = plt.subplots(figsize = (10,10))
for i in range(0,N):
    r.append(r[i]*r[0])
    phi.append(phi[i] + phi[0])
    x.append(r[-1]*np.cos(phi[-1]))
    y.append(r[-1]*np.sin(phi[-1]))
    ax.annotate(i+2,  (x[-1], y[-1]))

plt.plot(x,y, marker = 'o')
```
